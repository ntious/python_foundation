## Step 1: Store Daily Ticket Counts

Start with a few simple values. These numbers are small so the output is easy to check by hand.

```python
# TODO: Change these numbers to create a new test case.
monday_tickets = 4
tuesday_tickets = 6
wednesday_tickets = 3

print("Monday:", monday_tickets)
print("Tuesday:", tuesday_tickets)
print("Wednesday:", wednesday_tickets)
```

## Step 2: Store the Values in a List

A list keeps related values together. This makes it easier to process each value with a loop.

```python
# TODO: Add the three ticket variables to this list.
ticket_counts = [monday_tickets, tuesday_tickets, wednesday_tickets]

print(ticket_counts)
```

## Step 3: Process the Data

Use a loop to add each ticket count to the total. Read the loop out loud if it helps: for each count in the list, add it to the total.

```python
total_tickets = 0

# TODO: Complete the loop by adding each count to total_tickets.
for count in ticket_counts:
    total_tickets = total_tickets + count

print("Total tickets:", total_tickets)
```

## Step 4: Make a Decision

The help desk wants to know when the workload is high. Use an `if` statement to choose a message.

```python
high_workload_limit = 12

# TODO: Change the message text if you want it to be clearer.
if total_tickets > high_workload_limit:
    status_message = "High workload"
else:
    status_message = "Workload is manageable"

print(status_message)
```

## Step 5: Display the Final Output

A useful program prints results in a way another person can understand.

```python
# TODO: Add one more print line that gives advice to the help desk.
print("Help Desk Ticket Summary")
print("Ticket counts:", ticket_counts)
print("Total tickets:", total_tickets)
print("Status:", status_message)
```
