## Concept Introduction (no code first)

A list is a collection of related data. Instead of storing one value at a time, we can store many values together in a list.

## Creating a List

A list uses square brackets. Items are separated by commas. `len()` tells us how many items are in the list.

**Think Before You Run:** How many scores are in this list?

````python
# A list of related score values
scores = [88, 91, 85, 90]

# len() returns the number of items in the list.
print("Scores:", scores)
print("Number of scores:", len(scores))
````

## Accessing Elements

Each item has a position number called an index. List positions start at `0`, so the first item is at position `0`.

**Think Before You Run:** What value will `scores[0]` return?

````python
# List positions start at 0.
scores = [70, 80, 90]

print("First score:", scores[0])
print("Second score:", scores[1])
print("Third score:", scores[2])
````

## Think Before You Run

What happens if we use an index that does not exist?

````python
# Intentional error: there is no item at index 3.
# Run this once, read the message, then move on.
system_readings = [72, 75, 81]
print(system_readings[3])
````

## Looping Through a List

A loop can go through each item in a list one by one. This is helpful when we want to print, check, or calculate with every value.

````python
# The loop handles one login attempt value at a time.
login_attempts = [1, 3, 2, 5]

for attempts in login_attempts:
    print("Login attempts:", attempts)
````

## Loop Visualization

A loop moves through the list one item at a time:

```text
scores = [70, 80, 90]
score = 70 -> process
score = 80 -> process
score = 90 -> process
Loop ends
```

````python
# Watch the loop process each disk reading.
disk_readings = [72, 75, 81]

for reading in disk_readings:
    print("Checking reading:", reading)
````

## Data Processing (sum and average)

We are processing the data in the list. `sum()` adds the values, and `len()` counts how many items are in the list.

**Think Before You Run:** What total do you expect for these three readings?

````python
# Compute total and average.
readings = [70, 80, 90]

total = sum(readings)
average = total / len(readings)

print("Total:", total)
print("Average:", average)
````

## Real-World Examples

Lists are useful for small IT data tasks, such as system logs, usage metrics, login attempts, and monitoring values.

````python
# Disk usage monitoring
disk_usage_values = [68, 74, 82, 91]
threshold = 80

for usage in disk_usage_values:
    print("Disk usage:", usage)
    if usage > threshold:
        print("Warning: above threshold")

# Login attempt data
login_attempts = [1, 2, 1, 4]
average_attempts = sum(login_attempts) / len(login_attempts)
print("Average login attempts:", average_attempts)
````

## Common Mistakes

Common list mistakes include using an index that does not exist, forgetting the loop variable, or forgetting to start a total at `0`. Always check your list size and loop structure.

Index out of range:

```python
# Broken: this list has positions 0, 1, and 2 only.
values = [10, 20, 30]
print(values[3])
```

Forgetting the loop variable:

```python
# Fixed pattern
for value in values:
    print(value)
```

Missing initialization:

```python
# Start the total before the loop.
total = 0
```
