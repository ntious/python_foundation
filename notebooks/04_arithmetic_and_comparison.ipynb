## Concept Explanation

Arithmetic means doing math with values. Comparison means checking how two values relate to each other.

First, we use arithmetic to calculate values. Then we use comparison operators to check those values and make decisions.

## Arithmetic in Context

**Think Before You Run:** What total do you expect for the expenses below?

```python
# Total expenses using +
books = 45.50
software = 24.99
printing = 6.50

total_expenses = books + software + printing
print("Total expenses:", total_expenses)
```

**Think Before You Run:** What will this calculate for page count and average score?

## Order of Operations

Python follows the usual math order. Parentheses can make the order clear and change the result.

**Think Before You Run:** Will these two calculations give the same result?

```python
# Without parentheses, multiplication happens before addition.
result_one = 10 + 5 * 2

# Parentheses tell Python what to do first.
result_two = (10 + 5) * 2

print("Without parentheses:", result_one)
print("With parentheses:", result_two)
```

Parentheses change the order of calculation. This short example shows the difference clearly.

```python
# Multiplication happens before addition.
print(2 + 3 * 4)

# Parentheses happen first.
print((2 + 3) * 4)
```

## Comparison Operators

After calculating a value, we can compare it to a target or threshold. Comparison expressions return either `True` or `False`.

`True` and `False` are Boolean values.

**Think Before You Run:** Will this result be `True` or `False`?

```python
# A comparison gives a Boolean result: True or False.
score = 75
print(score > 50)
```

The output is `True` because `75` is greater than `50`.

**Think Before You Run:** Which line will print `True`, and which line will print `False`?

```python
# Side-by-side comparison
score = 60

print(score > 50)
print(score < 50)
```

`score > 50` is `True` because 60 is greater than 50. `score < 50` is `False` because 60 is not less than 50.

**Think Before You Run:** Which disk usage line will print `True`, and which line will print `False`?

```python
# Comparison operators
disk_usage = 82
warning_level = 80

print("Greater than:", disk_usage > warning_level)
print("Less than:", disk_usage < warning_level)
print("Equal to:", disk_usage == warning_level)
print("Not equal to:", disk_usage != warning_level)
```

## Important: `=` vs `==`

`=` is used to assign a value. `==` is used to compare values.

```python
# = means assignment: store 10 in the variable.
open_tickets = 10

# == means comparison: ask whether the value is 10.
print("Open tickets is 10:", open_tickets == 10)
```

```python
# Short equality example
score = 80
print(score == 80)
```

## Common Mistakes

Small typing mistakes are common. Check symbols carefully, especially `=` and `==`.

```python
# Mistake to avoid: using = when you meant to compare.
# This line stores a value.
status_code = 200

# This line compares a value.
print("Status code is OK:", status_code == 200)
```

```python
# Mistake to avoid: forgetting parentheses for an average.
score_1 = 80
score_2 = 90
score_3 = 100

wrong_average = score_1 + score_2 + score_3 / 3
correct_average = (score_1 + score_2 + score_3) / 3

print("Wrong average:", wrong_average)
print("Correct average:", correct_average)
```
