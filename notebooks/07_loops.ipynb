## Concept Introduction (no code first)

A loop allows a program to repeat a task multiple times instead of writing the same code again and again. Examples include repeated login attempts and repeated system checks.

## First Loop Example

**Think Before You Run:** How many times will the message print?

````python
# This loop repeats the indented line 3 times.
for count in range(3):
    print("System check complete")
````

## Understanding range()

`range(5)` means repeat 5 times. Keep this idea simple for now: the number inside `range()` tells Python how many repeats to make.

**Think Before You Run:** If `range(4)` is used, how many times will this run?

````python
# range(4) repeats 4 times.
for step in range(4):
    print("Checking login attempt")
````

## Loop Variable

The loop variable changes each time the loop runs. With `range(5)`, it changes from 0 to 4.

````python
# The loop variable is named attempt here.
for attempt in range(3):
    print("Attempt number:", attempt)
````

**Think Before You Run:** What is the first value of `i`?

````python
# i changes from 0 to 4.
for i in range(5):
    print(i)
````

## Think Before You Run

This version adds 1 so the display starts at 1 instead of 0. What numbers do you expect to see?

````python
# Add 1 when you want people-friendly numbering.
for attempt in range(3):
    print("Login attempt:", attempt + 1)
````

## Real-World Examples

Loops are useful when an IT task repeats. Examples include login attempts, system monitoring, and repeated system checks.

````python
# Example 1: login attempts
for attempt in range(3):
    print("Allowed login attempt:", attempt + 1)

# Example 2: system checks
for system_number in range(4):
    print("Checking system", system_number + 1)

# Example 3: repeated data entry reminder
for form_number in range(2):
    print("Enter data for form", form_number + 1)
````

## Loop Visualization

A loop follows a simple pattern:

```text
Start loop
Repeat 1: run the indented code
Repeat 2: run the indented code
Repeat 3: run the indented code
Loop ends
```

Each repeat is called an iteration, but you can think of it as one loop step.

Loop variable example:

```text
i = 0 -> print
i = 1 -> print
i = 2 -> print
...
```

`range(5)` produces `0, 1, 2, 3, 4` which is 5 values.

````python
# Watch the loop steps appear one at a time in the output.
for step in range(3):
    print("Loop step:", step + 1)

print("Loop finished")
````

## Common Mistakes

Python loops depend on indentation and correct `range()` values. Common mistakes include indentation errors, misunderstanding `range()`, and off-by-one errors.

Indentation mistake:

```python
# Broken
for item in range(3):
print("Check item")

# Fixed
for item in range(3):
    print("Check item")
```

Off-by-one reminder:

```python
# range(5) repeats 5 times, but the loop variable is 0, 1, 2, 3, 4.
```

````python
# Correct pattern to copy when you are unsure.
for number in range(3):
    print("Displayed number:", number + 1)
````

## Mini Lab

Use repeated input to collect three numbers and calculate a total. The input line repeats each time the loop runs.

````python
# Mini Lab: repeated input and total
# Step 1: Start the total at 0.
# Step 2: Repeat input 3 times.
# Step 3: Add each value to the total.
# Enter three small numbers, one at a time.

total = 0

for entry in range(3):
    value_text = input("Enter a number: ")
    value = float(value_text)
    # Add the new value to the running total.
    total = total + value

print("Total:", total)
````
