## Introducing Multiple Conditions

Some decisions need more than one check. For example, access may require both an ID card and an active account.

## AND Logic

`and` means both conditions must be `True`. If only one condition is `True`, the whole check is `False`.

**Think Before You Run:** The user has an ID card and an active account. Should access be allowed?

````python
# AND example: system access control
has_id_card = True
has_active_account = True

if has_id_card and has_active_account:
    print("Access allowed")
else:
    print("Access denied")
````

**Think Before You Run:** What happens if only one condition is `True`?

````python
# AND example: one condition is False
has_id_card = True
has_active_account = False

print("AND result:", has_id_card and has_active_account)
````

## OR Logic

`or` means at least one condition must be `True`. The whole check is `False` only when both conditions are `False`.

**Think Before You Run:** If either security warning is `True`, should the alert print?

````python
# OR example: security alert
many_failed_logins = False
unknown_location = True

if many_failed_logins or unknown_location:
    print("Security alert")
else:
    print("No alert")
````

**Think Before You Run:** What happens if both conditions are `False`?

````python
# OR example: both conditions are False
many_failed_logins = False
unknown_location = False

print("OR result:", many_failed_logins or unknown_location)
````

## Comparing AND vs OR

The same two values can give different results with `and` and `or`. `and` is stricter. `or` is more flexible.

**Think Before You Run:** With one `True` and one `False`, what will `and` do? What will `or` do?

````python
# Side-by-side comparison
has_network = True
has_admin_permission = False

print("AND result:", has_network and has_admin_permission)
print("OR result:", has_network or has_admin_permission)
````

**Think Before You Run:** Which line will be `True`, and which line will be `False`?

````python
# Side-by-side comparison with score and attendance
score = 60
attendance = 70

print(score >= 50 and attendance >= 75)
print(score >= 50 or attendance >= 75)
````

With `and`, both requirements must pass. With `or`, one passing requirement is enough.

Mini visual:

```text
AND:
True AND True -> True
True AND False -> False

OR:
True OR False -> True
False OR False -> False
```

## Real-World Examples

Multiple conditions help with IT decisions such as system access control, security alerts, login attempts, and resource monitoring.

````python
# Resource monitoring example
disk_usage = 86
memory_usage = 72

if disk_usage > 80 or memory_usage > 80:
    print("Resource warning")
else:
    print("Resources look okay")

# Access control example
completed_training = True
signed_policy = True

if completed_training and signed_policy:
    print("User can receive system access")
else:
    print("User must finish requirements first")
````

**Think Before You Run:** Which security condition causes the alert?

````python
# Security alert example
failed_login_attempts = 5
unusual_location = False

if failed_login_attempts > 3 or unusual_location:
    print("Security review needed")
else:
    print("No security review needed")
````

## Common Mistakes

Common mistakes include confusing `and` with `or`, assuming one `True` is enough for `and`, and writing overly complex conditions. Start simple. Test one condition at a time.

````python
# Mistake to avoid: using and when one warning should be enough.
low_storage = True
low_battery = False

# For alerts, or often makes more sense because either issue matters.
if low_storage or low_battery:
    print("Show warning")
else:
    print("No warning")
````
