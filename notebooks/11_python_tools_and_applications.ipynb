## Dictionaries (Structured Data)

A dictionary stores data as key-value pairs, like a record or profile. Dictionaries can represent a user profile, system configuration, or other structured data.

## Think Before You Run

What value do you expect Python to show for `status`?

```python
# A small dictionary with system information
system = {"name": "Lab Printer", "status": "ready"}

print(system["status"])
```

## Real-World Meaning of Dictionaries

Dictionaries are useful when one item has several details. For example, a device record can have a room and status.

**Think Before You Run:** What value will `device["room"]` return?

```python
# Access one detail by using its key.
device = {"room": "A204", "status": "online"}

print("Room:", device["room"])
```

## Introduction to Libraries (concept only)

A library is prewritten code that allows us to perform tasks without building everything from scratch. We import a library when we want to use what it provides.

## Library Demonstration (very short)

This example uses the `random` library to pick a simple support ticket priority.

**Think Before You Run:** What do you expect this to display?

```python
import random

priority = random.choice(["low", "medium", "high"])
print("Ticket priority:", priority)
```

## Think Before You Run

Run the random example more than once. Does the output always stay the same?

```python
import random

check_number = random.randint(1, 3)
print("Random check number:", check_number)
```
