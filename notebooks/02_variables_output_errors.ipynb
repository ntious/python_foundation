## What is a Variable?

A variable is like a labeled container that stores information your program can use later. The label is the variable name, and the stored information is the value.

**Think Before You Run:** What do you think this code will print for the username and score?

````python
# Variables store information for later use.
username = "alex"
score = 92

print("Username:", username)
print("Score:", score)
````

## Errors Are Normal

Errors are expected, useful, and part of learning Python. Each error message is a clue that can help you find one small thing to fix.

## Exploring Common Errors

The next examples are broken on purpose. Run the broken code, read the message, then compare it with the corrected version.

### NameError

**Think Before You Run:** Will this code run successfully if `server_status` was never created?

````python
# Broken on purpose: server_status has not been defined.
print("Server status:", server_status)
````

A `NameError` often means Python does not know that variable name. Create the variable before using it.

````python
# Corrected version
server_status = "online"
print("Server status:", server_status)
````

### SyntaxError

**Think Before You Run:** What small typing mistake do you notice in this print statement?

````python
# Broken on purpose: the closing quote is missing.
print("Disk check complete)
````

A `SyntaxError` means Python cannot understand the code as typed. Check quotes, parentheses, brackets, and colons first.

````python
# Corrected version
print("Disk check complete")
````

### TypeError

**Think Before You Run:** What error might happen when text and a number are combined with `+`?

````python
# Broken on purpose: this tries to add text and a number.
disk_message = "Disk usage is " + 72
print(disk_message)
````

A `TypeError` can happen when values do not work together. For beginner output, using commas in `print()` is a simple fix.

````python
# Corrected version
disk_usage = 72
print("Disk usage is", disk_usage)
````

## How to Read Errors

Start with the last line of the error message. Identify the error name, then check the line Python mentioned.

Use this simple process:

1. Look at the last line.
2. Find the error name.
3. Check the line mentioned.
4. Fix one small thing and run again.
