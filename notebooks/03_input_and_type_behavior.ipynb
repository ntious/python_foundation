## Concept Explanation (plain language, no jargon)

`input()` lets a program ask the user a question. The important rule is: **`input()` always returns text (a string), even if the user types a number**.

A string is text. If we want to do math with input, we usually need to convert it first.

## Demonstration

**Think Before You Run:** If you type `5`, do you think Python sees it as a number or as text?

````python
# Try typing a number, such as 5.
user_answer = input("Type a number: ")

# input() gives back text, even if the text looks like a number.
print("You typed:", user_answer)
print("Python sees this as:", type(user_answer))
````

## Intentional Error (students run code that fails)

This cell has a planned error:

Do NOT fix it yet. The next section shows the fix.

````python
# Intentional error: adding a string and an integer does not work.
login_attempts_text = "3"
total_attempts = login_attempts_text + 1
print(total_attempts)
````

## Fix and Explanation

Python cannot combine text and numbers directly. The fix is to convert the text first. Use `int()` when you want a whole number.

````python
# Convert the text "3" into the number 3.
login_attempts_text = "3"
login_attempts = int(login_attempts_text)

total_attempts = login_attempts + 1
print("Total attempts:", total_attempts)
````

**Think Before You Run:** What type will this value be before conversion? What type will it be after `int()`?

input() gives text, conversion turns it into numbers.

````python
# Short demo: raw input type compared with converted type
tickets_text = input("Enter number of tickets: ")
print("Raw input:", tickets_text)
print("Before conversion:", type(tickets_text))

tickets = int(tickets_text)
print("Converted number:", tickets)
print("After conversion:", type(tickets))
````

## Real-World Examples (IT-related)

Small IT tasks often include numbers that start as text. We convert them before comparing or calculating.

**Think Before You Run:** Which examples use `int()` and which use `float()`?

````python
# Example 1: disk usage
disk_percent_text = "82.5"
disk_percent = float(disk_percent_text)
print("Disk usage:", disk_percent, "%")

# Example 2: login attempts
attempts_text = "4"
attempts = int(attempts_text)
print("Login attempts after one more try:", attempts + 1)

# Example 3: system value
open_tickets_text = "7"
open_tickets = int(open_tickets_text)
print("Open tickets:", open_tickets)
````

## Error Handling (basic try-except only)

Sometimes users type something unexpected. A basic `try-except` is a safety net when users enter unexpected input.

**Think Before You Run:** What should happen if the user types a word instead of a number?

````python
# Try typing a number first, then run again and try typing a word.
age_text = input("Enter your age: ")

try:
    age = int(age_text)
    print("Next year you will be", age + 1)
except:
    print("Please enter a whole number next time.")
````
