### What is Computer Programming?
* Computer programming is the process of **giving clear, step-by-step instructions to a computer** so it can perform a task.
* A **program** is simply a set of instructions written in a language the computer can understand.

### Think About It
Programming is similar to:

* A **recipe** → step-by-step instructions to cook a meal
* A **GPS route** → directions to reach a destination
* A **checklist** → steps to complete a task

In all these cases:
* The steps must be **clear and in order**
* Missing or incorrect steps lead to **wrong results**

### Important Idea
Computers do **exactly what you tell them to do**.
* If your instructions are clear → the program works
* If your instructions are unclear or incorrect → you get an **error or unexpected output**
This is a normal part of programming.

### Why Programming Matters
Programming allows you to:
* Automate repetitive tasks
* Work with data (e.g., scores, expenses, logs)
* Make decisions using logic
* Solve simple real-world problems

## Demo 1: A Simple Real-World Example

Run the code and observe the output.


```python
# Example: counting support tickets

open_tickets = 4
new_tickets = 2
total_tickets = open_tickets + new_tickets

print("Open tickets:", open_tickets)
print("New tickets:", new_tickets)
print("Total tickets to handle:", total_tickets)
```
