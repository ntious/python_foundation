## Demo 1: A Simple Real-World Example

This short program keeps track of support tickets.  
Run the code and observe the output.


```python
# Example: counting support tickets

open_tickets = 4
new_tickets = 2
total_tickets = open_tickets + new_tickets

print("Open tickets:", open_tickets)
print("New tickets:", new_tickets)
print("Total tickets to handle:", total_tickets)
```

## Explain the Code

This program does three simple things:

1. stores values in variables
2. adds the values together
3. displays the result with `print()`

This is one way Python helps people solve practical problems.
