## Concept Introduction (no code first)

People make decisions all day: if it is raining, bring an umbrella. Programs can do the same kind of checking.

A program checks a condition. If the condition is `True`, one action happens. If it is `False`, a different action may happen.

## Single Condition (if)

`if` statements use conditions that evaluate to `True` or `False`. If the condition is `True`, the indented block runs. If the condition is `False`, the block is skipped.

## Think Before You Run

The disk usage is `85`. Will Python print the warning message? What will still print even if the condition is `False`?

````python
# Single condition example: disk usage warning
# This script checks the disk usage and prints a warning if it exceeds 80%.
# The condition inside an if statement must evaluate to True or False.
disk_usage = 85

if disk_usage > 80:
    print("Warning: disk usage is high")

print("Disk check complete")
````

## if-else

`if-else` gives Python two paths. With `if-else`, exactly one path will run.

## Think Before You Run

The account is active. Which message do you expect?

````python
# System access example
account_active = True

# If a variable already stores True or False, Python can check it directly.
if account_active:
    print("Access allowed")
else:
    print("Access denied")
````

## if-elif-else

`elif` means "else if." Python checks the conditions from top to bottom and runs the first matching block.

## Think Before You Run

There are `4` login attempts. Which message should appear?

````python
# Login attempts example
login_attempts = 4

if login_attempts < 3:
    print("Login activity looks normal")
elif login_attempts == 3:
    print("Warning: one more failed attempt may lock the account")
else:
    print("Account needs support")
````

## Flow Visualization

A decision works like a simple path:

1. Check the condition.
2. If `True`: run this block.
3. Otherwise: check the next option or use `else`.
4. Continue the program.

```text
Check condition
-> if True: run this block
-> otherwise: check next option or use else
-> continue program
```

## Real-World Examples

Decision-making appears in many IT tasks, such as checking access, storage, and login activity.

````python
# Example 1: system access
has_id_card = True

if has_id_card:
    print("User can enter the lab")
else:
    print("User needs an ID card")

# Example 2: disk usage threshold
disk_usage = 78

if disk_usage > 90:
    print("Critical disk warning")
elif disk_usage > 80:
    print("Disk warning")
else:
    print("Disk usage looks okay")
````

## Common Mistakes

These mistakes are common. Focus on what changed in the corrected version.

Missing colon:

```python
# Broken
if disk_usage > 80

# Fixed
if disk_usage > 80:
```

Wrong indentation:

```python
# Broken
if account_active:
print("Access allowed")

# Fixed
if account_active:
    print("Access allowed")
```

Using `=` instead of `==`:

```python
# Broken
if login_attempts = 3:

# Fixed
if login_attempts == 3:
```

If you get stuck, compare your code line by line with a working pattern.

````python
# Correct pattern to copy when you are unsure.
login_attempts = 3

if login_attempts == 3:
    print("Three attempts recorded")
````
