## Concept Introduction (no code first)

A function is a reusable block of code that performs a task. Functions help us organize programs and avoid repeating the same code.

## Repetition Problem

Without a function, repeated messages can make code longer than it needs to be. This code repeats the same task multiple times. Is there a better way?

**Think Before You Run:** What task is repeated in this example?

````python
# Repeated code without a function
print("Running system check")
print("System check complete")

print("Running system check")
print("System check complete")

print("Running system check")
print("System check complete")
````

## Defining a Function

`def` starts a function. Defining a function does not run it. A function runs only when it is called.

````python
# Define the function first.
def run_system_check():
    print("Running system check")
    print("System check complete")
````

**Think Before You Run:** Will anything happen if we define a function but do not call it?

## Calling a Function

Calling a function means telling Python to run it. Use the function name followed by parentheses.

````python
# Call the function to run its code.
run_system_check()
run_system_check()
````

## Think Before You Run

What will happen when we call this function three times?

````python
# A short function can be reused many times.
def show_login_message():
    print("Welcome to the course system")

show_login_message()
show_login_message()
show_login_message()
````

## Function with Parameter (simple)

A parameter allows us to pass information into a function. In this notebook, we will usually use only one parameter at a time.

````python
# device_name is one parameter.
def show_device_status(device_name):
    print(device_name, "is ready")

show_device_status("Printer")
show_device_status("Router")
````

## Real-World Examples

Functions are useful for repeated IT tasks, such as system checks, login messages, and simple data processing.

````python
# Example 1: system check
def check_system():
    print("System status: OK")

check_system()

# Example 2: login message with one parameter
def greet_user(user_name):
    print("Welcome,", user_name)

greet_user("Student")

# Example 3: simple data processing with one parameter
def show_next_attempt(attempts):
    print("Next attempt number:", attempts + 1)

show_next_attempt(3)
````

## Optional: Return Values (light introduction)

A return value is a result that the function gives back. We can store that result in a variable.

````python
# Light return example
def add(a, b):
    return a + b

total = add(2, 3)
print("Total:", total)
````

## Common Mistakes

Common function mistakes include forgetting parentheses, wrong indentation, or calling a function before it has been defined. Functions must be defined before they are used.

Forgetting parentheses:

```python
# This shows the function object, but does not run it.
show_login_message

# This runs the function.
show_login_message()
```

Indentation error:

```python
# Broken
def check_system():
print("System OK")

# Fixed
def check_system():
    print("System OK")
```

Calling before defining:

```python
# Define the function first, then call it.
```
